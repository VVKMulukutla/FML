```python
from sklearn.decomposition import PCA
import sklearn.metrics as metrics
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_breast_cancer

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
cancer_data = load_breast_cancer()
cancer = pd.DataFrame(cancer_data['data'], columns = cancer_data['feature_names'])
cancer
```

```python
cancer_data['target']
```

```python
sc = StandardScaler()
scaled_cancer = sc.fit_transform(cancer)
scaled_cancer = pd.DataFrame(scaled_cancer, columns = cancer_data['feature_names'])
scaled_cancer
```

```python
pca = PCA(n_components = 2)

new_pca = pca.fit_transform(scaled_cancer)

print(scaled_cancer.shape)
print(new_pca.shape)
```

```python
sns.scatterplot(x = new_pca[:, 0], y= new_pca[:, 1], c= cancer_data['target'], cmap = 'winter', marker = "X")
```
